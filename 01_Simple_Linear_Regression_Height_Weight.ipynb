```python
# !pip install pandas matplotlib scikit-learn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Load and view the data
df = pd.read_csv("https://raw.githubusercontent.com/jrue/ML-Examples-2026/refs/heads/main/human_sample.csv")
df.head()
```

```python
# Set the X and y for our model
X = df[["Height"]]
y = df["Weight"]

# Train the model
model = LinearRegression()
model.fit(X, y)

# Plot the data + best-fit line
plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Data")
y_pred = model.predict(X)
plt.plot(X, y_pred, label="Best Fit Line", color="red")

plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight (Linear Regression)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Interactive prediction
def predict_weight():
    while True:
        user_input = input("Enter a height: ")

        try:
            height = float(user_input)
            #prediction = model.predict([[height]])[0]
            sample = pd.DataFrame([[height]], columns=["Height"])
            prediction = model.predict(sample)[0]
            print(f"Predicted weight: {prediction:.2f}\n")
            break
        except:
            print("Please enter a valid number.\n")

predict_weight()
```

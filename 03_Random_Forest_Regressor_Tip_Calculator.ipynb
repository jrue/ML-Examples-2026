```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error
```

```python
# LOAD DATA
df = pd.read_csv("https://raw.githubusercontent.com/jrue/ML-Examples-2026/refs/heads/main/tips.csv")
df.head(100)
```

```python
df.describe()
```

```python
# set up features
features = ["total_bill", "size", "sex", "smoker", "day", "time"]
X = df[features].copy()
y = df["tip"]

# Convert categorical to numeric
encoders = {}
for col in ["sex", "smoker", "day", "time"]:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    encoders[col] = le
```

```python
# Train the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)
preds = model.predict(X_test)

print("Model accuracy:", model.score(X_test, y_test))
print("MAE:", mean_absolute_error(y_test, preds))
```

```python
# Feature importance (signal strength)
importances = model.feature_importances_

plt.figure()
plt.barh(features, importances)
plt.xlabel("Importance (Signal Strength)")
plt.title("Which Features Matter Most?")
plt.show()
```

```python
# Now your turn. Experiment.
def predict_tip():
    print("\nEnter a new example:")

    total_bill = float(input("Total bill: "))
    size = int(input("Party size: "))
    sex = input("Sex (Male/Female): ")
    smoker = input("Smoker (Yes/No): ")
    day = input("Day (Thur/Fri/Sat/Sun): ")
    time = input("Time (Lunch/Dinner): ")

    sample = pd.DataFrame([{
        "total_bill": total_bill,
        "size": size,
        "sex": encoders["sex"].transform([sex])[0],
        "smoker": encoders["smoker"].transform([smoker])[0],
        "day": encoders["day"].transform([day])[0],
        "time": encoders["time"].transform([time])[0],
    }])

    prediction = model.predict(sample)[0]

    print(f"\nPredicted tip: ${prediction:.2f}")

predict_tip()
```

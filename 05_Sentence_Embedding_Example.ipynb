```python
#pip install sentence-transformers scikit-learn matplotlib

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
# A small set of sentences with a few related meanings
sentences = [
    "The moon is made of cheese.",
    "The moon is made of rock.",
    "The moon orbits the Earth.",
    "Mars is a planet.",
    "Jupiter is a planet.",
    "Cheese is made from milk.",
    "Pizza has cheese on it.",
    "The Earth orbits the sun.",
    "Saturn has rings.",
    "Milk comes from cows."
]
```

```python
# Load a pretrained sentence embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Turn each sentence into an embedding vector
embeddings = model.encode(sentences)

for i, s in enumerate(sentences[:3]):
    print("\nSentence:", s)
    print("First 12 embedding values:", embeddings[i][:12])
```

```python
# Reduce many dimensions down to 2 so we can plot them
pca = PCA(n_components=2)
points_2d = pca.fit_transform(embeddings)

# Plot
plt.figure(figsize=(10, 7))
for i, (x, y) in enumerate(points_2d):
    plt.scatter(x, y)
    plt.text(x + 0.02, y + 0.02, str(i), fontsize=9)

plt.title("Sentence Embeddings Projected to 2D")
plt.xlabel("PCA dimension 1")
plt.ylabel("PCA dimension 2")
plt.grid(True)
plt.show()

print("\nLegend:")
for i, s in enumerate(sentences):
    print(f"{i}: {s}")
```

```python
# Compute similarity
similarity_matrix = cosine_similarity(embeddings)

# Plot heatmap
plt.figure(figsize=(10, 8))
plt.imshow(similarity_matrix, cmap="Blues", vmin=-0, vmax=1)
plt.colorbar(label="Cosine Similarity")

# Add labels
plt.xticks(range(len(sentences)), range(len(sentences)), rotation=90)
plt.yticks(range(len(sentences)), range(len(sentences)))

plt.title("Sentence Similarity (Cosine Similarity Heatmap)")
plt.xlabel("Sentence Index")
plt.ylabel("Sentence Index")

for i in range(len(sentences)):
    for j in range(len(sentences)):
        plt.text(j, i, f"{similarity_matrix[i, j]:.2f}",
                 ha="center", va="center", color="white", fontsize=6)

plt.tight_layout()
plt.show()

# Print legend
print("\nLegend:")
for i, s in enumerate(sentences):
    print(f"{i}: {s}")
```
